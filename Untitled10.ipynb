```python
from sklearn.cluster import KMeans
```

```python
k_means = KMeans(n_clusters=3, max_iter=10000)
```

```python
df=df.dropna()
```

```python
k_means.fit(df[['total_cases_per_million', 'total_deaths_per_million']])
```

```python
centers = k_means.cluster_centers_

centers
```

```python
clasificaciones = k_means.predict(df[['total_cases_per_million', 'total_deaths_per_million']])
```

```python
fig = plt.figure()
ax = fig.add_subplot()

ax.set_title('Relación de muertes por covid vs número de casos', pad=15)
ax.set_xlabel('Número de casos por millón')
ax.set_ylabel('Número de muertes por millón')

sns.scatterplot(df['total_cases_per_million'], df['total_deaths_per_million'], ax=ax, hue=clasificaciones, palette='rainbow');
sns.scatterplot(centers[:,0], centers[:,1], ax=ax, s=100, color='black');

ax.get_legend().remove()
```

```python
asia = merged[merged['continent'] == 'Asia']
africa = merged[merged['continent'] == 'Africa']
europe = merged[merged['continent'] == 'Europe']
southamerica = merged[merged['continent'] == 'South America']
northamerica = merged[merged['continent'] == 'North America']
oceania = merged[merged['continent'] == 'Oceania']
```

```python
sns.set_style('whitegrid')
palette = ["#223127","#9c0d38","#ce5374","#dbbbf5","#F9A03F",'green','blue']

fig = plt.figure(figsize=(7, 7))
axes = fig.subplots(3, 2, sharex='col', sharey='row')

sns.scatterplot(asia['total_cases_per_million'], asia['total_deaths_per_million'], ax=axes[0, 0], color=palette[4])
sns.scatterplot(africa['total_cases_per_million'], africa['total_deaths_per_million'], ax=axes[0, 1], color=palette[1])
sns.scatterplot(europe['total_cases_per_million'], europe['total_deaths_per_million'], ax=axes[1, 0], color=palette[2])
sns.scatterplot(southamerica['total_cases_per_million'], southamerica['total_deaths_per_million'], ax=axes[1, 1], color=palette[3]);
sns.scatterplot(northamerica['total_cases_per_million'], northamerica['total_deaths_per_million'], ax=axes[2, 0], color=palette[5]);
sns.scatterplot(oceania['total_cases_per_million'], oceania['total_deaths_per_million'], ax=axes[2, 1], color=palette[0]);

axes[0, 0].set_title('Asia')
axes[0, 1].set_title('Africa')
axes[1, 0].set_title('Europe')
axes[1, 1].set_title('South America')
axes[2, 0].set_title('North America')
axes[2, 1].set_title('Oceania')

fig.suptitle('Relationship between Total Cases and Total Deaths per million in the different Continents', fontsize=15, c=palette[0], x=0.53)

plt.tight_layout()
plt.subplots_adjust(top=0.9)
```
