```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style('white')
from google.colab import drive
drive.mount('/content/drive')
```

```python
df = pd.read_csv('/content/drive/MyDrive/covid/edad.csv')

df.head()
```

```python
df['age_group'].value_counts()
```

```python
counts=df['age_group'].value_counts()
```

```python
fig, ax = plt.subplots(figsize=(30,10))
ax = sns.barplot(counts.index, counts)
ax.set_title('Conteo de Información de vacunados grupo de Edades')
ax.set(ylabel='count');
```

```python
as_percentages = counts * 100 / counts.sum()
```

```python
fig, ax = plt.subplots(figsize=(30,10))
ax = sns.barplot(as_percentages.index, as_percentages)
ax.set_title('Conteo de Información de personas vacunadas por edad(como porcentajes)')
ax.set(ylabel='porcentaje del total');
```

```python
df = pd.read_csv('/content/drive/MyDrive/covid/paises.csv', index_col=0)

df.head()
```

```python
by_continent = df.groupby('continent')[['total_cases_per_million', 'total_deaths_per_million', 'population_density','cardiovasc_death_rate','diabetes_prevalence','life_expectancy','hdi']].mean()
```

```python
by_continent
```

```python
fig, ax = plt.subplots(figsize=(10,8))
plt.suptitle('')
df.boxplot(column=['total_deaths_per_million'], by='continent', ax=ax)
```

```python
grouped = df.groupby(level=0)[['continent']].last()
```

```python
grouped.head()
```

```python
by_country = df.groupby(level=0)[['total_cases_per_million', 'total_deaths_per_million', 'population_density','cardiovasc_death_rate','diabetes_prevalence','life_expectancy','hdi']].mean()
```

```python
by_country
```

```python
merged = by_country.merge(grouped, left_index=True, right_index=True)
```

```python
sns.violinplot(data=merged, x='continent', y='total_deaths_per_million')
```

```python
sns.boxplot(data=merged, x='continent', y='total_deaths_per_million')
```
